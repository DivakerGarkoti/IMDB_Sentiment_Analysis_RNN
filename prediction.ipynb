#### loading Dataset

```python
from tensorflow.keras.datasets import imdb
```

```python
max_features=10000
(X_train,Y_train),(x_test,y_test)=imdb.load_data(num_words=max_features)
```

```python
sample_review=X_train[0]
```

```python
word_index=imdb.get_word_index()
```

```python
reverse_word_index={(values,keys) for keys, values in word_index.items()}
```

```python
reverse_word_index
```

```python
from tensorflow.keras.models import load_model
model=load_model("model.h5")
```

```python
model.summary()
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing import sequence
def decode_review(encoded_review):
    return ' '.join([reverse_word_index.get(i-3, '?') for i in encoded_review])

def preprocess_text(text):
    words=text.lower().split()
    encoded_review=[word_index.get(word,2)+3 for word in words]
    padded_review=sequence.pad_sequences([encoded_review],maxlen=500)
    return padded_review
```

```python
def predict_sentiment(review):
    preprocessed_text = preprocess_text(review)
    prediction = model.predict(preprocessed_text)
    sentiment = 'Positive' if prediction[0][0]>0.5 else 'Negative'
    return prediction[0][0], sentiment
```

```python
example_review="This movie was fantastic! The acting was greate and the plot was thrilling"
```

```python
prediction,sentiment=predict_sentiment(example_review)
print("review:",example_review)
print("Sentiment:",sentiment)
print('Prediction:',prediction)
```
